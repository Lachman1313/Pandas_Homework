```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_group = purchase_data["Gender"].value_counts()
total_gender = gender_group.sum()
perc_gender = gender_group/total_gender *100
demo_df = pd.concat([perc_gender, gender_group], axis=1)
demo_df.columns = ["Percentage of Players", "Total Counts"]
demo_df["Percentage of Players"] = demo_df["Percentage of Players"].map("{:,.2f}%".format)
demo_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gen_group = purchase_data.groupby(["Gender"])
pur_price = gen_group["Price"].sum()
avg_price = pur_price / gender_group

genpur_df = pd.concat([gender_group, avg_price, pur_price], axis=1)
genpur_df.columns = ["Purchase Count", "Average Purchase Price", "Total Purchases"]
genpur_df["Total Purchases"] = genpur_df["Total Purchases"].map("${:,.2f}".format)
genpur_df["Average Purchase Price"] = genpur_df["Average Purchase Price"].map("${:,.2f}".format)
genpur_df.index.name = "Gender"

genpur_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bin = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 1000]
age_grouping = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Group"]=pd.cut(purchase_data["Age"], age_bin, labels=age_grouping)
age_grp_df = purchase_data.groupby(["Age Group"])


age_purch = age_grp_df["Age"].count()
age_tot = age_grp_df["Price"].sum()
age_avg = age_tot / age_purch
big_tot = age_tot.sum()
age_pct = age_tot / big_tot * 100

age_based_df = pd.concat([age_purch, age_avg, age_tot, age_pct], axis=1)
age_based_df.columns = ["Purchase Count", "Average Purchase", "Total Purchases", "Percent of Total Value"]
age_based_df["Average Purchase"] = age_based_df["Average Purchase"].map("${:,.2f}".format)
age_based_df["Total Purchases"] = age_based_df["Total Purchases"].map("${:,.2f}".format)
age_based_df["Percent of Total Value"] = age_based_df["Percent of Total Value"].map("{:,.2f}".format)

age_based_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
user_df = purchase_data.groupby(["SN"])
users = user_df["Purchase ID"].count()
user_tot = user_df["Price"].sum()
user_avg = user_tot / users

top_use_df = pd.concat([users, user_tot, user_avg], axis=1)
top_use_df.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
top_use_df = top_use_df.sort_values("Purchase Count", ascending=False)
top_use_df["Total Purchase Value"] = top_use_df["Total Purchase Value"].map("${:,.2f}".format)
top_use_df["Average Purchase Price"] = top_use_df["Average Purchase Price"].map("${:,.2f}".format)

top_use_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_items = purchase_data[["Item ID", "Item Name", "Price"]]
top_items_grp = top_items.groupby(["Item ID", "Item Name"])

top_count = top_items_grp["Item ID"].count()
top_totals = top_items_grp["Price"].sum()
top_price = top_totals / top_count

top_df = pd.concat([top_count, top_price, top_totals], axis=1)
top_df.columns = ["Purchase Count", "Item Price", "Total Value"]

org_top_df = top_df.sort_values("Purchase Count", ascending=False)
org_top_df["Item Price"] = org_top_df["Item Price"].map("${:,.2f}".format)
org_top_df["Total Value"] = org_top_df["Total Value"].map("${:,.2f}".format)

org_top_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profits_df = top_df.sort_values("Total Value", ascending=False)
profits_df["Item Price"] = profits_df["Item Price"].map("${:,.2f}".format)
profits_df["Total Value"] = profits_df["Total Value"].map("${:,.2f}".format)

profits_df.head()
```
